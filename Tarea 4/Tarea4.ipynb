```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img= cv2.imread('cameraman.png', cv2.IMREAD_GRAYSCALE)
gray = cv2.normalize(img.astype('float'), None, 0.0, 1.0, cv2.NORM_MINMAX)
F = np.fft.fft2(gray)
vector = np.linspace(-0.5, 0.5, gray.shape[0])
U,V = np.meshgrid(vector, vector)
a = 5
b = 1
UV = U*a+V*b
G = F*np.fft.fftshift(np.sinc(np.pi*UV)*np.exp(-1j*np.pi*UV))
g = np.real(np.fft.ifft2(G))
plt.imshow(g, cmap='gray')
plt.show()

```

```python
#2. (50%)
#1. Restaure la imagen del siguiente código empleando el filtro de Wiener y el filtro Paramétrico.
#Determine el parámetro que mejor resultado genere.
```

```python
#Cargamos las librerias
import cv2
import numpy as np
import numpy.matlib
from matplotlib import cm

Test = np.fft.fft2(g)
Hf = np.fft.fftshift(np.sinc(np.pi*UV)*np.exp(-1j*np.pi*UV))
K =1 #Asignamos un valor a nuestro parametro,
W = np.conj(Hf)/(np.abs(Hf)**2 + K) #aplicamos la ecuación del filtro
F = W*Test #multiplicamos la formula por la imagen degradada.
iRestored = np.real(np.fft.ifft2(F))

plt.imshow(iRestored, cmap='gray') #imprimimos por pantalla
plt.show()
```

```python
#Reiteramos los pasos
K =10
W = np.conj(Hf)/(np.abs(Hf)**2 + K)
F = W*Test
iRestored = np.real(np.fft.ifft2(F))

plt.imshow(iRestored, cmap='gray')
plt.show()
```
