```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
#Tarea 2 - Benjamin Parraguez
#Pregunta 1

img = cv2.imread('fallas.tif',0) #0 porque trabajamos en escala de grises.
plt.imshow(img, cmap = 'gray')
plt.show()
```

```python
#Pregunta 2

ImgEcual = cv2.equalizeHist(img)
plt.imshow(ImgEcual, cmap = 'gray')
plt.show()
```

```python
#Como podemos apreciar, ambas imagenes son iguales, no obstante, se ven diferentes, debido a que
#la segunda imagen esta ecualizada y por ende, se aprecia un mayor contraste, además
#podemos notar que se pueden apreciar de mejor manera las fallas a causa de este fenomeno.
```

```python
#Pregunta 3

Mediana_test1 = cv2.medianBlur(ImgEcual, 11)
Mediana_test2 = cv2.medianBlur(ImgEcual, 19)
Mediana_test3 = cv2.medianBlur(ImgEcual, 23)
Mediana_test4 = cv2.medianBlur(ImgEcual, 33)

plt.imshow(Mediana_test1, cmap='gray')
plt.show()

plt.imshow(Mediana_test2, cmap='gray')
plt.show()

plt.imshow(Mediana_test3, cmap='gray')
plt.show()

plt.imshow(Mediana_test4, cmap='gray')
plt.show()
```

```python
#Luego de manipular y observar los posibles valores que atenuan las fallas, he llegado a conclusión que
#el tamaño de la máscara más "óptima" , debe poseer un valor de 33.
```

```python
#Pregunta 4 

Resta_test1 = cv2.subtract(Mediana_test1,ImgEcual) #11
Resta_test2 = cv2.subtract(Mediana_test2,ImgEcual) #19
Resta_test3 = cv2.subtract(Mediana_test3,ImgEcual) #23
Resta_test4 = cv2.subtract(Mediana_test4,ImgEcual) #33

plt.imshow(Resta_test1, cmap='gray')
plt.show()

plt.imshow(Resta_test2, cmap='gray')
plt.show()

plt.imshow(Resta_test3, cmap='gray')
plt.show()

plt.imshow(Resta_test4, cmap='gray')
plt.show()
```

```python
#Comparando los resultados dados en la resta, podemos notar que a mayor valor que toma la máscara,
#los fallos comienzan a desaparecer, caso parecido cuando los valores que uno escoge son muy bajos.
```

```python
#Pregunta 5
ret, bina_test = cv2.threshold(Resta_test4, 35, 255, type = cv2.THRESH_BINARY)

plt.imshow(bina_test, cmap="gray")
plt.show()
```

```python
#Luego de probar valores entre el umbral de 0 a 255, he decidido que se aprecia de mejor forma con un
#valor de 35.
```

```python
#Pregunta 6

Multi_BinaEc= bina_test * ImgEcual
plt.imshow(Multi_BinaEc, cmap= 'gray')
plt.show()
```
